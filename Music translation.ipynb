```python
from music21 import stream, note
import random

melody = stream.Stream()

notes = ['C', 'D', 'E', 'F', 'G', 'A', 'B']

for _ in range(20):
    n = note.Note(random.choice(notes))
    n.quarterLength = 0.5
    melody.append(n)

melody.write('midi', 'generated_music.mid')
melody
```
